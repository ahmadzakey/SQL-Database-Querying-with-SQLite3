```python
import sqlite3
import pandas as pd
```

```python
# Sambung ke SQLite database
conn = sqlite3.connect('Data/movie.sqlite')

# Query untuk mendapatkan senarai jadual dalam database
df = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", conn)

# Papar senarai jadual
print("Senarai Jadual dalam Database:")
df
```

# Table IDMB

```python
imdb = "SELECT * FROM imdb"

df_imdb = pd.read_sql_query(imdb, conn)

print("Data dalam jadual 'imdb':")
df_imdb.head()
```

# Table genre

```python
genre = "SELECT * FROM genre"

df_genre = pd.read_sql_query(genre, conn)

print("Data dalam jadual 'genre':")
df_genre.head()
```

# Table Earning

```python
earning = "SELECT * FROM earning"

df_earning = pd.read_sql_query(earning, conn)

print("Data dalam jadual 'earning':")
df_earning.head()
```

## PANDAS MANIPULATION

## JOIN 3 Table with valuable column for next analsis

```python
# Gabungkan df_imdb dan df_genre berdasarkan Movie_id
merged_df_1 = pd.merge(df_imdb, df_genre, on='Movie_id', how='inner')

# Gabungkan hasil dari merged_df_1 dengan df_earning berdasarkan Movie_id
merged_df_final = pd.merge(merged_df_1, df_earning, on='Movie_id', how='inner')

# Gabungkan genre untuk setiap Movie_id
merged_df_final = merged_df_final.groupby(['Movie_id', 'Title', 'Rating', 'TotalVotes', 'MetaCritic', 'Budget', 'Runtime', 'CVotesMale', 'CVotesFemale', 'Domestic', 'Worldwide'], as_index=False)['genre'].apply(lambda x: ', '.join(x)).reset_index(drop=True)

# Paparkan hasilnya
merged_df_final[['Movie_id', 'Title', 'Rating', 'TotalVotes', 'MetaCritic', 'Budget', 'Runtime', 'genre', 'CVotesMale', 'CVotesFemale', 'Domestic', 'Worldwide']]
```

```python
# Simpan merged_df_final ke dalam fail CSV
merged_df_final.to_csv('merged_movie_data.csv', index=False)

# Paparkan mesej berjaya
print("Data berjaya disimpan dalam fail 'merged_movie_data.csv'")
```
